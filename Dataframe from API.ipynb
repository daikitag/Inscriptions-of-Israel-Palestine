```python
import pandas as pd
import xml.etree.ElementTree as ET
import urllib.request

# Choose your own url
url = 'https://library.brown.edu/search/solr_pub/iip/?start=0&rows=5000&indent=on&wt=xml&q=*'

dataf={'IIP ID':['Summary','Text Genre','City Name','Language','Terminus ante quem','Terminus post quem','Object Type',
               'Region','Likely religion','City Name','Description of Place Found','Material','Original text','Translation',
               'Description']}

response = urllib.request.urlopen(url).read()
rot = ET.fromstring(response)
rot=rot.find('result')

for i in range (0,5000):
    try:
        [ID,summary,typ,city,language,notBefore,notAfter,objec,region,religion,city,detail_place,material,original,
               translation,description]=['','','','','','','','','','','','','','','','']

        root=rot[i]

        # IIP ID
        try:
            ID=root.find('str[@name="inscription_id"]').text
        except:
            pass

        # Summary
        try:
            summary=root.find('str[@name="short_description"]').text
            summary=summary.replace('\n','')
        except:
            pass

        # Text Genre
        try:
            typ = root.find('arr[@name="type"]')[0].text
        except:
            pass

        # City Name
        try:
            city=root.find('str[@name="city"]').text
        except:
            pass

        # Language
        try:
            language =root.find('arr[@name="language_display"]')[0].text
        except:
            pass
        
        # Terminus ante quem
        try:
            notBefore = root.find('int[@name="notBefore"]').text
        except:
            pass
        
        # Terminus post quem
        try:
            notAfter=root.find('int[@name="notAfter"]').text
        except:
            pass
        
        # Object Type
        try:
            objec=root.find('arr[@name="physical_type"]')[0].text
        except:
            pass
        
        # Region
        try:
            region=root.find('str[@name="region"]').text
        except:
            pass
        
        # Likely religion
        try:
            religion=root.find('arr[@name="religion"]')[0].text
        except:
            pass
        
        # City Name
        try:
            city=root.find('str[@name="city"]').text
        except:
            pass
        
        # Description of Place Found
        try:
            detail_place=root.find('str[@name="place_found"]').text
            detail_place=detail_place.replace('\n                           ','')
            detail_place=detail_place.replace('\n','')
        except:
            pass

        # Material
        try:
            material=root.find('arr[@name="material"]')[0].text
        except:
            pass

        # Original text
        try:
            original=root.find('arr[@name="text"]')[0].text
            original=original.replace('\n','')
        except:
            pass
        
        # Translation
        try:
            translation=root.find('arr[@name="text"]')[1].text
            translation=translation.replace('\n','')
        except:
            pass

        # Description
        try:
            description=root.find('str[@name="description"]').text
            description=description.replace('\n','')
        except:
            pass

        dataf[ID]=[summary,typ,city,language,notBefore,notAfter,objec,region,religion,city,detail_place,material,original,
               translation,description]
    except:
        pass

df = pd.DataFrame(dataf)
df.to_csv('summary.csv', index=False)
```
